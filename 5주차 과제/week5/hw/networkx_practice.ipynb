```python
import networkx as nx
```

```python
# 노드와 엣지 만들기
G=nx.Graph()

print(G.nodes())
print(G.edges())

print(type(G.nodes()))
print(type(G.edges()))
```

```python
# 노드 추가 (개별 value, list로 추가 가능)
G=nx.Graph()

# adding just one node:
G.add_node("a")
# a list of nodes:
G.add_nodes_from(["b","c"])

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
G=nx.Graph()

# 노드 추가
G.add_node("a")
G.add_nodes_from(["b","c"])

#방법1. add_edge함수의 인자로 직접 입력
G.add_edge(1,2)

#방법2. edge 변수에 두개의 노드를 입력 후, add_edge의 인자로 *edge 입력
edge = ("d", "e")
G.add_edge(*edge)
edge = ("a", "b")
G.add_edge(*edge)

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
# adding a list of edges:
G.add_edges_from([("a","c"),("c","d"), ("a",1), (1,"d"), ("a",2)])
print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G, with_labels = True)
#plt.savefig("simple_path.png") # save as png
plt.show() # display|
```

```python
G=nx.path_graph(4)

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
nx.draw(G,with_labels = True)
#plt.savefig("path_graph1.png")
plt.show()
```

```python
G = nx.Graph() # 원하는 파라미터 이름 = 실수 (**kwds) 
G.add_edge(1, 2, weight=5)
G.add_edge(2, 3, weight=3)
G.add_edge(3, 4, weight=2)
G.add_edge(2, 4, weight=1)
G.add_edge(1, 4, weight=0.5)
# 아래는 그래프로 그리는 법 
pos=nx.spring_layout(G) 
nx.draw(G, pos=pos, with_labels=True)
# edge 라벨은 weight 값을 가져온다. 
labels = nx.get_edge_attributes(G,'weight') 
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels) 

# 이 py 파일 위치하는 곳에 G.png 저장 
#plt.savefig("G.png") 
plt.show()
```
